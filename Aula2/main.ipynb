```python
import pandas as pd

tabela = pd.read_csv("cancelamentos.csv")
display(tabela)
```

## Tratando os Dados
Precisamos fazer uma limpeza na tabela para que não tenhamos dados em excesso.   
A coluna CustomerID não será relevante para nossa análise de dados, por isso vamos apagar

```python
tabela = tabela.drop("CustomerID", axis=1)
display(tabela)
```

```python
display(tabela.info())
```

Ao exibir as informações da tabela `display(tabela.info())` podemos observar que nem todas as linhas possuem os dados completos. 
Vamos utilizar o comando `dropna` para remover as informações vazias da tabela

```python
tabela = tabela.dropna()
display(tabela.info())
```

## Análise da taxa de cancelamento
Com a base de dados tratada, iniciaremos a análise verificando a taxa de cancelamento, tendo em vista que o objetivo é diminuir essa taxa

```python
# Verificando a quantidade de pessoas que cancelaram
display(tabela["cancelou"].value_counts())
```

```python
# Colocando em porcentagem
display(tabela["cancelou"].value_counts(normalize=True))
```

```python
# Alterando visualização
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

## Verificando a Duração do Contrato

```python
display(tabela["duracao_contrato"].value_counts())
```

```python
display(tabela["duracao_contrato"].value_counts(normalize=True))
```

## Analisando as Informações dos Contratos 

```python
# analisando contrato mensal
display(tabela.groupby("duracao_contrato").mean(numeric_only=True))
```

Com as informações agrupadas, é possível notar que os clientes do plano Mensal, possuem uma média de cancelamento igual a 1, ou seja, praticamente todos os clientes que utilizam esse plano fizeram o cancelamento do serviço.

## Removendo o contrato mensal
Sabendo que o contrato mensal é ruim para a empresa, podemos remover as informações desse contrato específico e continuar analisando.

```python
tabela = tabela[tabela["duracao_contrato"]!="Monthly"]
display(tabela)
```

```python
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

Após remover os assinantes mensais, podemos observar que a taxa de cancelamento caiu para 46.1%, porém, ainda precisamos avançar na análise

## Análise de Assinaturas

```python
display(tabela["assinatura"].value_counts(normalize=True))
display(tabela.groupby("assinatura").mean(numeric_only=True))
```

Podemos observar que a quantidade de assinantes de Standard / premium / basic é bem parecida e que a taxa de cancelamento também é parecida.
